```python
for directors, cast in df.values:
    directors = str(directors)
    cast = str(cast)
    if directors == 'nan' or directors == np.nan:
        pass
    else:
        directors = [d.strip() for d in directors.split(',')]
        directors = list(dict.fromkeys(directors))
        for d in directors:
            print(d)
    if cast == 'nan' or cast == np.nan:
        pass
    else:
        cast = [c.strip() for c in cast.split(',')]
        cast = list(dict.fromkeys(cast))
        for actor in cast:
            print(actor)
        
```
